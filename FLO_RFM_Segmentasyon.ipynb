```python
rfm = df.groupby("master_id").agg({"Last_order_date" : lambda Last_order_date: (today_date - Last_order_date.max()).days,
                                  "total_number_of_purchases" : lambda total_number_of_purchases : total_number_of_purchases.sum(),
                                  "customer_value_total_ever" : lambda customer_value_total_ever : customer_value_total_ever.sum()})
```

```python
rfm.head()
```

```python
rfm.shape
```

```python
rfm.columns = ["Recency", "Frequency", "Monetary"]
```

```python
rfm.head()
```

```python
rfm.dtypes
```

```python
rfm.describe().T
```

```python
#Recency, Frequency ve Monetary metriklerinin 1-5 arasÄ±nda skorlara Ã§evirmesi.
```

```python
rfm["RecencyScore"] = pd.qcut(rfm["Recency"], 5, labels = [5, 4, 3 , 2, 1])
```

```python
rfm["FrequencyScore"] = pd.qcut(rfm["Frequency"].rank(method="first"), 5, labels= [1, 2, 3, 4, 5])
```

```python
rfm["MonetaryScore"] = pd.qcut(rfm["Monetary"], 5, labels = [1, 2, 3, 4, 5])
```

```python
rfm.head()
```

```python
#RF skorunun belirlenmesi
```

```python
rfm["RFM_Score"] = (rfm['RecencyScore'].astype(str) +
                    rfm['FrequencyScore'].astype(str))
```

```python
rfm.head()
```

```python
#RF skorlarÄ± iÃ§in segment tanÄ±mlamalarÄ±
```

```python
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At_Risk',
    r'[1-2]5': 'Cant_Loose',
    r'3[1-2]': 'About_to_Sleep',
    r'33': 'Need_Attention',
    r'[3-4][4-5]': 'Loyal_Customers',
    r'41': 'Promising',
    r'51': 'New_Customers',
    r'[4-5][2-3]': 'Potential_Loyalists',
    r'5[4-5]': 'Champions'
}
```

```python
rfm["Segment"]= rfm["RFM_Score"].replace(seg_map, regex =True)
```
